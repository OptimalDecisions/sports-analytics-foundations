## 1. Anatomy of a Matplotlib Figure


To understand Matplotlib, it helps to know its underlying hierarchy.



### 1.1 Figure

At the very top, a `Figure` object is a container for all of the elements that make up our final image.

It controls foundational settings like the `size` of our final image, or whether we want a `legend`.

Inside of the figure, there are one or more `axes`. (I personally feel that this is a very unfortunate and confusing name.)

-  Within a figure, we have one or more subplots, each represented by an Axes object.



### 1.2 Axes

An `axes` is essentially a plot.
, and it holds the information for what we are graphing.

If we wanted to have several visualizations, then we would need several axes.

 - An Axes is an individual plot within the figure and contains elements such as the x-axis, y-axis, labels, and the actual plot.

As can be seen from the "Anatomy" image above, there are a huge number of options. You don't have to memorize these, just know that they exist. We can look them up as needed.




### 1.3 High-Level Components

- `Figure`: The top-level container that holds all elements of a plot.
- `Axes`: The individual plot within a figure.
- `plt.subplots()`: A function to create a figure and one or more subplots.
- `ax.plot()`: Plots data on a specific subplot.
- `ax.set_title()`, `ax.set_xlabel()`, `ax.set_ylabel()`: Customize appearance.
- `ax.legend()`: Add a legend to distinguish between different lines or data.



```python
import matplotlib.pyplot as plt
```

### 1.4 Create a Basic Plot

Use  `plt.plot()` to create a basic plot.


```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y = [2, 4, 4, 6, 10]

# Create a Figure and Axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, label='Line Plot')

# Customize the appearance
ax.set_title('Simple Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.show()
```

Note that even in this simple plot, we customized the appearance with `ax.set_title()`, `ax.set_xlabel()`, `ax.set_ylabel()`, and `ax.legend()`.



### 1.5. Customizing the Appearance

```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

# Create a Figure and Axes
fig, ax = plt.subplots()

# Plot the data with different styles and colors
ax.plot(x, y1, label='Line 1', linestyle='-', color='blue', marker='o')

# Add a second line to the same plot (same ax)
ax.plot(x, y2, label='Line 2', linestyle='--', color='green', marker='s')

# Customize the appearance
ax.set_title('Customized Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.show()
```

We have now introduced a few new elements, such as   `linestyle`, `color`, and `marker` to customize the appearance of the lines.

### 1.6 Mutliple Plots in the same figure

Often, we want to create two subplots side by side within the same figure.

Matplotlib is ideal for doing that. This is very easy to do, thanks to the `plt.subplots()` command. It lets us design the number of rows and columns for our plots. For ex. 2x1 or 1x2 plots (1 row, 2 columns).


```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

# Create a Figure with two subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# Plot data on the first subplot
ax1.plot(x, y1, label='Line 1', linestyle='-', color='blue', marker='o')
ax1.set_title('Subplot 1')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

# Plot data on the second subplot
ax2.plot(x, y2, label='Line 2', linestyle='--', color='green', marker='s')
ax2.set_title('Subplot 2')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
